### Yelp

```python
dfYelpBusiness.sample(5)
```

```python
dfYelpBusiness[dfYelpBusiness['len_list']==16]
```

```python
# craer dataset
dfYelpBusiness.to_parquet('YelpBusiness.parquet', index=False)
```

# Reviews

---------------

### Yelp

```python
# crear lista con business_id datasets YelpBusiness
listYelp = []
listYelp = dfYelpBusiness['business_id'].to_list()
```

```python
len(listYelp)
```

```python
dfYelpReviews = pd.read_parquet('/content/drive/MyDrive/DataYelpGoogle/Yelp/reviews.parquet', engine='fastparquet')
```

```python
dfYelpReviews.info()
```

```python
dfYelpReviews.head(5)
```

```python
dfYelpReviews.rename({'stars': 'rating'}, axis=1, inplace=True)
```

```python
dfYelpReviews = dfYelpReviews[['review_id',	'user_id',	'business_id',	'rating',	'text']]
```

```python
dfYelpReviews.head(5)
```

```python
dfYelpReviewsAPI = pd.read_parquet('/content/drive/MyDrive/DataYelpGoogle/Yelp/Yelp_Reviews_API.parquet')
```

```python
dfYelpReviewsAPI.info()
```

```python
dfYelpReviewsAPI.head(5)
```

```python
# armar un nuevo dataframe df
dfYelpReviews = pd.concat([dfYelpReviews, dfYelpReviewsAPI], axis=0, ignore_index=True)
```

```python
dfYelpReviews.info()
```

```python
# eliminar duplicados
dfYelpReviews.drop_duplicates(subset=['review_id'], inplace=True)
dfYelpReviews.reset_index(drop=True, inplace=True)
```

```python
# numero de reviews donde coinciden los business filtrados
dfYelpReviews = dfYelpReviews[dfYelpReviews['business_id'].apply(lambda x: x in listYelp)]
dfYelpReviews.reset_index(drop=True, inplace=True)
```

```python
dfYelpReviews.info()
```

```python
dfYelpReviews.sample(5)
```

```python
# craer dataset
dfYelpReviews.to_parquet('YelpReviews.parquet', index=False)
```
