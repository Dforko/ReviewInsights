### Preprocesing

```python
# visualizar 1 registro
df['review'][0]
```

```python
# normalizar, limpiar texto - review
import re

def CleanText(textA):
    textB = textA.replace("\n", " ") # saltos de linea
    textC = textB.lower() # minusculas
    CleanedText = re.sub('[^a-zA-Z0-9]', ' ', textC) # caracteres especiales
    return CleanedText
```

```python
# clean
df['review processed'] = df['review'].apply(CleanText)
```

```python
# validar cambios
df['review processed'][0]
```

```python
# token
from nltk.tokenize import RegexpTokenizer

regExp = RegexpTokenizer('\w+')
```

```python
# token
df['review processed'] = df['review processed'].apply(regExp.tokenize)
```

```python
# validar cambios
df['review processed'][0]
```

```python
# stopwords
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords = nltk.corpus.stopwords.words('english')
```

```python
# stopwords
df['review processed'] = df['review processed'].apply(lambda x: [item for item in x if item not in stopwords])
```

```python
# validar cambios
df['review processed'][0]
```

```python
# delete words < 3 letters
df['review processed'] = df['review processed'].apply(lambda x: [item for item in x if len(item)>=3])
```

```python
# validar cambios
df['review processed'][0]
```

```python
# lenght review
df['lenght review'] = df['review processed'].apply(lambda x: len(x))
```

### Cluster

```python
# datos de entrenamiento
X_train = df['review processed']
```

```python
# Generating bag of words features.
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

X_train_tfidf = vectorizer.fit_transform(X_train)
```

```python
# kMeans
from sklearn.cluster import KMeans

k = 2
clu_model = KMeans(n_clusters=k, n_init='auto', max_iter=300, init='k-means++')
clu_model.fit(X_train_tfidf)
```

```python
# labels cluster
df['cluster'] = clu_model.labels_
```

```python
# show data
df.sample(5)
```

```python
# palabras mas representativas de cada cluster
print('Cluster centroids: \n')
order_centroids = clu_model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()

for i in range(k):
  print('Cluster %d:' %i)
  for j in order_centroids[i, :25]:
    print('%s' %terms[j])
  print('---------------------')
```

```python
# registros por cada dataset de entrada
df['dataset'].value_counts()
```

```python
### cluster objetivo puede variar entre 0 y 1 dependiendo del momento en el cual se ejecute el codigo, importante verificar en el momento de cada ejecucion
```

```python
# identificar reviews con menos de 5 palabras en cluster 'conocedores de vino'
df[(df['lenght review'] < 5) & (df['cluster'] == 0)]
```

```python
# cambiar de cluster objetivo al otro cluster las reviews con menos de 5 palabras
df['cluster'][df[(df['lenght review'] < 5) & (df['cluster'] == 0)].index] = 1
```

```python
# lemmatization
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import spacy

import nltk
nltk.download('wordnet')

nlp = spacy.load("en_core_web_sm")

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(text):
  pos = nlp(text)[0].pos_
  if pos == 'VERB':
    return 'v'
  elif pos == 'ADJ':
    return 'a'
  elif pos == 'ADV':
    return 'r'
  elif pos == 'NOUN':
    return 'n'
  else:
    return 'n'

lemmatizer = WordNetLemmatizer()
def lemmatization(text):
  return lemmatizer.lemmatize(text, get_wordnet_pos(text))
```
