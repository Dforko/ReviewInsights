### wine magazin

```python
dfWineMagazinReviews = pd.read_csv('/content/drive/MyDrive/winemag-data-130k-v2.csv')
```

```python
# info data
dfWineMagazinReviews.info()
```

```python
#  show data
dfWineMagazinReviews.head(3)
```

```python
# filtrar US
dfWineMagazinReviews = dfWineMagazinReviews[dfWineMagazinReviews['country']=='US']
dfWineMagazinReviews.reset_index(drop=True, inplace=True)
```

```python
# filtrar not America in province - America no nos especifica un estado, se decide eliminar para evitar confusiones
dfWineMagazinReviews = dfWineMagazinReviews[dfWineMagazinReviews['province']!='America']
dfWineMagazinReviews.reset_index(drop=True, inplace=True)
```

```python
# obtener columnas relevantes
dfWineMagazinReviews = dfWineMagazinReviews[['winery', 'province', 'taster_name', 'points', 'description']]
```

```python
# eliminar duplicados
dfWineMagazinReviews.drop_duplicates(inplace=True)
dfWineMagazinReviews.reset_index(drop=True, inplace=True)
```

```python
# filtrar notnull taster_name
dfWineMagazinReviews = dfWineMagazinReviews[dfWineMagazinReviews['taster_name'].notna()]
dfWineMagazinReviews.reset_index(drop=True, inplace=True)
```

```python
# renombrar
dfWineMagazinReviews.rename({'winery': 'business_name',	'province': 'state', 'taster_name': 'user_name', 'points': 'rating', 'description': 'review'}, axis=1, inplace=True)
```

```python
# info data
dfWineMagazinReviews.info()
```

```python
#  show data
dfWineMagazinReviews.sample(5)
```

```python
# crear columna review_id
dfWineMagazinReviews['review_id'] = [x for x in range(0, 34395)]
```

```python
dfWineMagazinReviews['review_id'] = dfWineMagazinReviews['review_id'].apply(lambda x: 'WM-' + str(x))
```

```python
# codificar variables tipo object a valor numerico - user name
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
```

```python
dfWineMagazinReviews['user_name'] = le.fit_transform(dfWineMagazinReviews['user_name'])
dfWineMagazinReviews['business_id'] = le.fit_transform(dfWineMagazinReviews['business_name'])
```

```python
dfWineMagazinReviews['user_name'] = dfWineMagazinReviews['user_name'].apply(lambda x: 'WM-' + str(x))
dfWineMagazinReviews['business_id'] = dfWineMagazinReviews['business_id'].apply(lambda x: 'WM-' + str(x))
```

```python
# renombrar
dfWineMagazinReviews.rename({'user_name': 'user_id'}, axis=1, inplace=True)
```

```python
# cambiar rango rating de 80-100 a consideracion del autor 3-5
dfWineMagazinReviews['rating'] = dfWineMagazinReviews['rating'].apply(lambda x: (x<85 and 3) or (x<95 and 4) or 5)
```

```python
# crear columna categories
dfWineMagazinReviews['categories'] = 'winery, vineyard'
```

```python
dfWineMagazinReviews['categories'] = dfWineMagazinReviews['categories'].str.split(pat=',')
```

```python
# obtener columnas relevantes
dfWineMagazinReviews = dfWineMagazinReviews[['review_id', 'business_id', 'business_name', 'categories', 'state', 'user_id', 'rating', 'review']]
```

```python
dfWineMagazinReviews.info()
```

```python
#  show data
dfWineMagazinReviews.sample(5)
```

```python
dfWineMagazinReviews['dataset'] = 'WineMagazin'
```

```python
dfWineMagazinReviews.sample(5)
```
