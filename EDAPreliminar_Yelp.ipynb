-------------------

```python
df = pd.read_pickle("/content/drive/MyDrive/DataYelpGoogle/business.pkl")
```

```python
# info data
df.info()
```

```python
# show data
df.sample(5)
```

probelma al leer archivo .pkl las columnas se duplican, por lo cual se hara un nuevo archivo .parquet en el cual se tendra en cuenta las priemras columnas y se tomara 5 registros validos de las demas

```python
# df1  columna 0 a columna 13
df1 = df.iloc[:, 0:14]
```

```python
# df2 columnas 14 en adelante
df2 = df.iloc[:, 14:28]
```

```python
# extraer registros validos de las columnas a descartar
df2 = df2[df2['business_id'].notnull()]
```

```python
# armar un nuevo dataframe
df = pd.concat([df1, df2], axis=0)
```

```python
# info data
df.info()
```

```python
# cambiar tipo de datos
df.latitude = df.latitude.astype('float64')
df.longitude = df.longitude.astype('float64')
df.stars = df.stars.astype('float32')
df.review_count = df.review_count.astype('int32')
```

```python
# show data
df.sample(5)
```

```python
# crear parquet columnas validas
df.to_parquet('business.parquet', index=False)
```
