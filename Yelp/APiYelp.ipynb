```python
import time
import pandas as pd
from gql import gql, Client
from gql.transport.requests import RequestsHTTPTransport
```

```python
# define api_key, define endpoint, define header M
API_key = API_key = API_keys.API_key
endPoint = 'https://api.yelp.com/v3/graphql'
header = {'Authorization': 'bearer %s' % API_key, 'Content-Type': 'application/json'}
```

```python
# Select your transport with a defined url endpoint
transport = RequestsHTTPTransport(url=endPoint, headers=header, use_json=True)
```

```python
# Create a GraphQL client using the defined transport
client = Client(transport=transport, fetch_schema_from_transport=True)
```

```python
# Provide a GraphQL query
offset=0
query = gql( """
 {
    search(categories: "pubs", location: "Olympia", limit:1, offset: %s, radius: 25000) {
        total
            business {
                id
                name
                alias
                categories {
                    title
                    alias
                }
                location {
                    address1
                    city
                    state
                    country
                }
                coordinates {
                    latitude
                    longitude
                }
                price
                rating
                review_count
                reviews(limit: 3) {
                    id
                    text
                    rating
                    user {
                        id
                        name
                    }
                }
            }
    }
}
""" % offset)
```

```python
# Execute the query on the transport
response = client.execute(query)
```

```python
response
```

```python
df = pd.DataFrame.from_dict(response['search']['business'], orient='columns')
```

# GRAPH QL

```python
# define api_key, define endpoint, define header M
API_key = API_key = API_keys.API_key
endPoint = 'https://api.yelp.com/v3/graphql'
header = {'Authorization': 'bearer %s' % API_key, 'Content-Type': 'application/json'}
```

```python
# Select your transport with a defined url endpoint
transport = RequestsHTTPTransport(url=endPoint, headers=header, use_json=True)
```

```python
# Create a GraphQL client using the defined transport
client = Client(transport=transport, fetch_schema_from_transport=True)
```

```python
# Provide a GraphQL query

offset=0
total_response=[]

for i in range(0, 50, 25):

  query = gql( """
  {
      search(categories: "wineries,winetastingroom", location: "California Carpinteria", limit:25, offset: %s, radius: 40000) {
          total
              business {
                  id
                  name
                  alias
                  categories {
                      title
                      alias
                  }
                  location {
                      address1
                      city
                      state
                      country
                  }
                  coordinates {
                      latitude
                      longitude
                  }
                  price
                  rating
                  review_count
                  reviews(limit: 3) {
                      id
                      text
                      rating
                      user {
                          id
                          name
                      }
                  }
              }
      }
  }
  """ % offset)

  offset+=25
  time.sleep(1)

  response = client.execute(query)

  total_response+=response['search']['business']
```
